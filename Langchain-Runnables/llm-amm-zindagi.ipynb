```python
import random
class FakeLLM:
    def __init__(self):
        print('LLM created')
    def predict(self, prompt):
        response_list = ['Delhi is the capital of India.', 
                        'IPL is a cricket league in India.', 
                        'AI stands for Artificial Intelligence.']
        return {'response': random.choice(response_list)}
```

```python
class FakePromptTemplate:
    def __init__(self, template, input_variables):
        self.template = template
        self.input_variables = input_variables
    def format(self, kwargs):
        return self.template.format(**kwargs)
```

```python
template = FakePromptTemplate(template='Write a {length} note about {topic}?', 
                            input_variables=['length', 'topic'])
```

```python
prompt = template.format(
    {
        'topic': 'India', 
        'length': 'short'
    }
    )
print(prompt)
```

```python
llm = FakeLLM()
```

```python
llm.predict(prompt)
```

```python
class FakeLLMChain():
    def __init__(self, llm: FakeLLM, prompt: FakePromptTemplate):
        self.llm = llm
        self.prompt = prompt
        
    def run(self, kwargs):
        formatted_prompt = self.prompt.format(kwargs)
        result = self.llm.predict(formatted_prompt)
        return result['response']
```

```python
template = FakePromptTemplate(template='Write a {length} note about {topic}?', 
                            input_variables=['length', 'topic'])
```

```python
llm = FakeLLM()
```

```python
chain = FakeLLMChain(llm=llm, prompt=template)
```

```python
chain.run({'length': 'short', 'topic': 'India'})
```
